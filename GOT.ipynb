# Game of Thrones Visualization
Data Source: https://github.com/jeffreylancaster/game-of-thrones/tree/master/data

```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.pyplot import figure
import seaborn as sns


def seconds(time):
    h, m, s = time.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)

# Import data from json file
with open('episodes.json') as f:
    data = json.load(f)

df = pd.DataFrame(columns = ['Season', 'Location', 'Time Spent'])

# For each scene in each episode, record the location, time spent, and season number
for episode in data['episodes']:
    season = episode['seasonNum']
    for scenes in episode['scenes']:      
        loc = scenes['location']
        time = seconds(scenes['sceneEnd']) - seconds(scenes['sceneStart'])
        if df.loc[(df.Location==loc) & (df.Season==season)].empty:
            df = df.append({'Season': season, 'Location': loc, 'Time Spent': time}, ignore_index = True)
        else: # Entry containing location and season is already in df.
            df.loc[(df.Location==loc) & (df.Season==season), 'Time Spent'] += time
            
            
pivot_df = df.pivot(index='Location', columns='Season', values='Time Spent')
# Set size of plot
plt.rcParams["figure.figsize"] = [12,12]
plot = pivot_df.plot.barh(stacked = True, colormap='RdBu')
plt.xlabel('Time in Seconds')
plt.title('Time Spent at Each Location Through the Seasons')
plt.show()
```
